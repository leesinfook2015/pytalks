## SciPy package

### Interpolation

```python
from scipy import interpolate
```

```python
def f(x):
    return np.sin(x) - 0.3 * x
```

```python
randn = np.random.randn

n_m = 10
x_m = np.arange(n_m)
y_m = f(x_m) + 0.15*randn(n_m) # simulated measured signal
```

```python
x_real = np.linspace(0, n_m-1, 100)
y_real = f(x_real)

lin_interp = interpolate.interp1d(x_m, y_m)  # return linear interpolation function
y_est_lin = lin_interp(x_real)   # evaluate at desired points

cub_interp = interpolate.interp1d(x_m, y_m, kind='cubic')  # return cubic interpolation function
y_est_cub = cub_interp(x_real)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x_real, y_real, 'k:', label='actual')
ax.plot(x_m, y_m, 'bo', label='measured')
ax.plot(x_real, y_est_lin, 'g-', label='linear')
ax.plot(x_real, y_est_cub, 'r-', label='cubic')
plt.legend(loc='lower left')
```

### Least-squares fit

```python
from scipy import stats
```

```python
n_pts = 50
x = np.linspace(-10, 20, n_pts)

a0, b0 = 0.75, -3

y = a0 * x + b0

y_noise = y + randn(n_pts)*2
```

```python
resu = stats.linregress(x, y_noise)
a, b = resu[0:2]

y_fit = a*x + b

eq_lbl = '$y = %7.2f x %+7.2f$' % (a,b)

plt.plot(x, y_noise, 'ko')
plt.plot(x, y_fit, 'b-', label=eq_lbl)
plt.legend(loc='upper left')
```

### Polynomial fit

```python
a0, b0, c0 = 0.5, -1.4, 8.0

y = a0*x**2 + b0*x + c0

y_noise = y + randn(n_pts)*10

plt.plot(x, y_noise, 'ko', mfc='none')  # mfc <-> markerfacecolor

a, b, c = np.polyfit(x, y_noise, 2)

eq_lbl = r'$%.2f x^2 %+.2f x %+.2f$' % (a, b, c)

y_fit = polyval((a,b,c), x)

plot(x, y_fit, 'b-', label=eq_lbl, lw=2)
legend(loc='upper left')
```

## scikit-learn

```python
from sklearn import datasets
```

```python
# generate some clustered data (X == features, L == labels)
X, L = datasets.make_blobs(centers=4, cluster_std=0.5, random_state=2)
plt.scatter(X[:,0], X[:,1])
```

```python
plt.scatter(X[:,0], X[:,1], c=L);
```

```python
from sklearn.cluster import KMeans
km = KMeans(4)
km.fit(X);
```

```python
plt.scatter(X[:,0], X[:,1], c=km.labels_);
mu = km.cluster_centers_
print(mu)
```
